# testing functional operation

```python
import math


def generate_vector(N=261121, step = 1/1024):
    out = [0]
    
    for i in range(N):
        #print(f"Appending: {out[i] + step}")
        out.append(out[i] + step)
    return out

def test(xs):
    total = 0
    for x in xs:
        total += 0.5 * x + x * x * math.cos((x - 128) / 128)
    return total
        

print(test([5, 20, -60, -20, 180, -60]))

x = generate_vector()
print(test(x))

    
```
