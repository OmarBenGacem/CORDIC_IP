```python
# Your array of data containing dictionaries
def moving_average_filter(data, n=10):
    filtered_data = []
    for i in range(len(data)):
        if i < n - 1:
            # For the first few elements where we don't have enough history
            filtered_data.append(sum(data[:i+1]) / (i + 1))
        else:
            # For the rest of the elements
            filtered_data.append(sum(data[i-n+1:i+1]) / n)
    return filtered_data
                
                
                
# Extracting the errors and steps from the dictionaries

steps = [entry["steps"] for entry in data]
cordic_errors = moving_average_filter([entry['cordic_error'] for entry in data])
taylor_errors = moving_average_filter([entry["taylor_error"] for entry in data])

# Plotting the errors against steps
plt.plot(steps, cordic_errors, label="16 Depth Cordic Data Error")
plt.plot(steps, taylor_errors, label="6th Degree Taylor Series Errors")
plt.axhline(y=0.5e-6, color='r', linestyle='--', label='Accuracy Threshold')
# Adding labels and title
plt.xlabel("Vector Length")
plt.ylabel("Mean Error")
plt.title("Mean Error of Calculated Cosine Values compared to Double Precision")
plt.legend()

# Displaying the plot
plt.grid(True)
plt.show()
```
